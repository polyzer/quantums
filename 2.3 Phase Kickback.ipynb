```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector
from qiskit_textbook.tools import array_to_latex
from math import pi
```

```python
## 1. Exploring the CNOT-Gate
```

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.draw()
```

```python
backend = Aer.get_backend('statevector_simulator')
res = execute(qc, backend).result().get_statevector()
array_to_latex(res, pretext = "\\text = {Result = }")
```

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.cx(0,1)
qc.draw()

statevector_backend = Aer.get_backend('statevector_simulator')
unitary_backend = Aer.get_backend('unitary_simulator')
final_state = execute(qc,statevector_backend).result().get_statevector()
array_to_latex(final_state, pretext="\\text{Statevector} = ", precision=1)
plot_bloch_multivector(final_state)
```

## 2. Phase Kickback

```python
qc = QuantumCircuit(2)
qc.cu1(pi/4, 0, 1)
qc.draw()
```

```python
qc = QuantumCircuit(2)
qc.cu1(pi/4, 0, 1)
display(qc.draw())
unitary_backend = Aer.get_backend('unitary_simulator')
unitary = execute(qc,unitary_backend).result().get_unitary()
array_to_latex(unitary, pretext="\\text{Controlled-T} = \n")
```
